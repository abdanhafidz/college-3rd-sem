### Latihan Soal
1. Tentukan apakah (~P & (P ==> Q)) ==> ~Q adalah tautologi?

```python
dfs = []
tautologi = True
for p in [True, False]:
    for q in [True, False]:
        prop1 = not p
        prop2 = not q
        prop3 = q if p else True
        prop4 = prop1 and prop3
        prop5 = prop2 if prop4 else True
        if(not prop5):
            tautologi = False
        dfs.append(pd.DataFrame({'P': [p], 
                                 'Q': [q], 
                                 '~P': [prop1], 
                                 '~Q':[prop2],
                                 'P ==> Q':[prop3],
                                 '~P & (P ==> Q)': [prop4],
                                 '~P & (P ==> Q) ==> ~Q' : [prop5]
                                }))

tb_soal_1 = pd.concat(dfs, ignore_index=True)

if(tautologi):
    print("Merupakan Tautologi")
else:
     print("Bukan Merupakan Tautologi")
```

```python
tb_soal_1
```

2. Tentukan apakah P <=> Q dan (p & q) | (~p & ~q) adalah ekivalen?

```python
dfs = []
ekivalen = True
for p in [True, False]:
    for q in [True, False]:
        prop1 = not p
        prop2 = not q
        prop3 = p and q
        prop4 = prop1 and prop2
        prop5 = prop3 or prop4
        prop6 = (q if p) and (p if q else True)
        prop7 = (prop5 if prop6 else True) and (prop6 if prop5 else True)
        if(not prop7):
            ekivalen = False
        dfs.append(pd.DataFrame({'P': [p], 
                                 'Q': [q], 
                                 '~P': [prop1], 
                                 '~Q':[prop2],
                                 'P & Q':[prop3],
                                 '~P & ~Q': [prop4],
                                 '(p & q) | (~p & ~q)' : [prop5],
                                 'p <=> q' : [prop6],
                                 '((p & q) | (~p & ~q)) <=> (p <=> q)':[prop7]
                                }))
tb_soal_2 = pd.concat(dfs, ignore_index=True)

if(ekivalen):
    print("Merupakan Ekivalen")
else:
     print("Bukan Merupakan Ekivalen")
```

```python
tb_soal_2
```

3. SUS banget

Di kelas ada 3 orang yang mungkin saja seorang SUS bernama Ajib, Layka, dan Zamtum yang hanya dapat berbohong atau jujur.

- Layka berkata : "Aku bukan orang SUS, Zamtumlah orang SUS"
- Zamtum : "Ajib adalah orang SUS"
- Ajib : "Aku jujur dan bukan orang SUS, Layka dan Zamtum pembohong!"


Dari ketiga pernyataan tadi dengan uji tabel kebenaran tentukanlah siapa orang yang dapat dipastikan SUS (orang yang berbohong ) di antara ketiganya! (orang sus mungkin saja >1)

Pembahasan

Kita dapat membuat proposisi 

- L = menyatakan bahwa Layka jujur
- Z = menyatakan bahwa Zamtum jujur
- A = menyatakan bahwa Azib jujur.


masing - masing L,Z, dan A dapat bernilai TRUE/FALSE.

Dari proposisi di atas kita dapat menggunakan kaidah interferensi untuk menentukan bahwa
- Jika Layka jujur maka Zamtum berbohong dan berlaku sebaliknya : (L <=> ~Z)
- Jika Zamtum jujur maka Layka berbohong dan Ajib juga berbohong : (Z <=> ~L & ~A)
- Jika Ajib jujur maka Layka dan Zamtum berbohong: (A <=> ~L & ~Z)

Kita bisa membuat uji dengan tabel kebenaran

```python
dfs = []
for L in [True, False]:
    for Z in [True, False]:
        for A in [True,False]:
            prop1 = not L
            prop2 = not Z
            prop3 = not A
            prop4 = prop1 and prop2
            prop5 = prop1 and prop3
            prop6 = (prop2 if L else True) and (L if prop2 else True)
            prop7 = (Z if prop5 else True ) and (prop5 if Z else True)
            prop8 = (A if prop4 else True ) and (prop4 if A else True)
            dfs.append(pd.DataFrame({'L': [L], 
                                    'Z': [Z], 
                                    'A': [A], 
                                    '~L':[prop1],
                                    '~Z':[prop2],
                                    '~A':[prop3],
                                    '~L & ~Z':[prop4],
                                    '~L & ~A': [prop5],
                                    'L <=> ~Z' : [prop6],
                                    'Z <=> ~L & ~A' : [prop7],
                                    'A <=> ~L & ~Z':[prop8]
                                    }))
tb_soal_3 = pd.concat(dfs, ignore_index=True)
```

```python
tb_soal_3
```

Periksa baris yang mengandung kolom L <=> ~Z, Z <=> ~L & ~A, dan A <=> ~L & ~Z bernilai benar

```python
tb_soal_3[tb_soal_3["L"]==True][tb_soal_3["L <=> ~Z"]==True][tb_soal_3["Z <=> ~L & ~A"]==True][tb_soal_3["A <=> ~L & ~Z"]==True]
```

Misalkan Zamtum Jujur

```python
tb_soal_3[tb_soal_3["Z"]==True]
```

```python
tb_soal_3[tb_soal_3["Z"]==True][tb_soal_3["L <=> ~Z"]==True][tb_soal_3["Z <=> ~L & ~A"]==True][tb_soal_3["A <=> ~L & ~Z"]==True]
```
