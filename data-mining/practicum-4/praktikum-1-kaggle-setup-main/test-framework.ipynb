```python
# Mengganti string kosong atau spasi dengan NaN
test_cleaned["winddirection_10m (°)"] = test_cleaned["winddirection_10m (°)"].replace(["", " ", "None"], pd.NA)
test_cleaned["winddirection_100m (°)"] = test_cleaned["winddirection_100m (°)"].replace(["", " ", "None"], pd.NA)

# Lanjutkan proses drop seperti sebelumnya
test_cleaned = test_cleaned[~(test_cleaned["winddirection_10m (°)"].isna() & test_cleaned["winddirection_100m (°)"].isna())]
```

```python
test_cleaned.isna().sum()
```

```python
from sklearn.linear_model import LinearRegression

# Fitur untuk regresi
features_10m = [
    "windspeed_10m (km/h)", 
    "pressure_msl (hPa)", 
    "temperature_2m (°C)", 
    "relativehumidity_2m (%)", 
    "apparent_temperature (°C)", 
    "vapor_pressure_deficit (kPa)", 
    "cloudcover (%)", 
    "shortwave_radiation (W/m²)"
]

features_100m = [
    "windspeed_100m (km/h)", 
    "pressure_msl (hPa)", 
    "temperature_2m (°C)", 
    "relativehumidity_2m (%)", 
    "apparent_temperature (°C)", 
    "vapor_pressure_deficit (kPa)", 
    "cloudcover (%)", 
    "shortwave_radiation (W/m²)"
]

# Imputasi untuk winddirection_10m (°)
# Data lengkap untuk pelatihan model winddirection_10m (°)
test_data_10m = test_cleaned.dropna(subset=["winddirection_10m (°)"])
missing_data_10m = test_cleaned[test_cleaned["winddirection_10m (°)"].isna()]

# Model regresi untuk winddirection_10m (°)
model_10m = LinearRegression()
model_10m.fit(test_data_10m[features_10m], test_data_10m["winddirection_10m (°)"])

# Prediksi untuk nilai yang hilang pada winddirection_10m (°)
test_cleaned.loc[test_cleaned["winddirection_10m (°)"].isna(), "winddirection_10m (°)"] = model_10m.predict(missing_data_10m[features_10m])

# Lakukan interpolasi linier atau forward fill jika masih ada nilai NaN
test_cleaned["winddirection_10m (°)"] = test_cleaned["winddirection_10m (°)"].interpolate(method='linear')

# Imputasi untuk winddirection_100m (°)
# Data lengkap untuk pelatihan model winddirection_100m (°)
test_data_100m = test_cleaned.dropna(subset=["winddirection_100m (°)"])
missing_data_100m = test_cleaned[test_cleaned["winddirection_100m (°)"].isna()]

# Model regresi untuk winddirection_100m (°)
model_100m = LinearRegression()
model_100m.fit(test_data_100m[features_100m], test_data_100m["winddirection_100m (°)"])

# Prediksi untuk nilai yang hilang pada winddirection_100m (°)
test_cleaned.loc[test_cleaned["winddirection_100m (°)"].isna(), "winddirection_100m (°)"] = model_100m.predict(missing_data_100m[features_100m])

# Lakukan interpolasi linier atau forward fill jika masih ada nilai NaN
test_cleaned["winddirection_100m (°)"] = test_cleaned["winddirection_100m (°)"].interpolate(method='linear')

# Cek jika ada nilai NaN yang tersisa
print("Jumlah NaN winddirection_10m (°):", test_cleaned["winddirection_10m (°)"].isna().sum())
print("Jumlah NaN winddirection_100m (°):", test_cleaned["winddirection_100m (°)"].isna().sum())
```

```python
test_cleaned.isnull().sum()
```

```python
# Mengubah kolom time menjadi tipe datetime
test_cleaned['time'] = pd.to_datetime(test_cleaned['time'])

# Memecah kolom time menjadi kolom terpisah
test_cleaned['year'] = test_cleaned['time'].dt.year
test_cleaned['month'] = test_cleaned['time'].dt.month
test_cleaned['day'] = test_cleaned['time'].dt.day
test_cleaned['hour'] = test_cleaned['time'].dt.hour
test_cleaned['minute'] = test_cleaned['time'].dt.minute
test_cleaned['second'] = test_cleaned['time'].dt.second

# Menggabungkan jam, menit, dan detik menjadi satu integer
# Menggunakan format jammenitdetik dalam bentuk integer
test_cleaned['time_int'] = test_cleaned['hour'] * 10000 + test_cleaned['minute'] * 100 + test_cleaned['second']

# Hapus kolom 'time' jika sudah tidak diperlukan
test_cleaned.drop(labels=['time','hour','minute','second'],axis=1,inplace=True)

# Menampilkan data hasil pemisahan
test_cleaned.info()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Label encoding untuk kolom 'city'
label_encoder = LabelEncoder()
test_cleaned['city'] = label_encoder.fit_transform(test_cleaned['city'])

# Mendefinisikan fitur yang ingin dinormalisasi, kecuali 'apparent_temperature (°C)' dan 'city'
features_to_scale = test_cleaned.columns.difference(['city', 'apparent_temperature (°C)'])

# StandardScaler untuk normalisasi data
scaler = StandardScaler()

# Melakukan scaling hanya pada fitur numerikal yang terpilih
test_cleaned[features_to_scale] = scaler.fit_transform(test_cleaned[features_to_scale])

# Menampilkan hasi
test_cleaned.head()

```
